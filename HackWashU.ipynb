```python
import random
import turtle
import time
from PIL import Image

delay = 0.1
score = 0

screen = turtle.Screen()
screen.title("Just Keep...eating")
screen.bgcolor("blue")
screen.setup(width = 600, height = 600)
screen.tracer(0)


# head of the snake
head = turtle.Turtle()
head.shape("triangle")
head.color("orange")
head.penup()
head.goto(0, 0)
head.direction = "Stop"

food = turtle.Turtle()
trash = turtle.Turtle()

shapes = random.choice(['square', 'triangle', 'circle'])
colors = random.choice(['pink', 'purple', 'black','grey','red'])

food.speed(0)
food.shape(shapes)
food.color(colors)
food.penup()
food.goto(0, 150)

trash.speed(0)
trash.shape(shapes)
trash.color(colors)
trash.penup()
trash.goto(0, 100)


pen = turtle.Turtle()
pen.speed(0)
pen.shape("square")
pen.color("white")
pen.penup()
pen.hideturtle()
pen.goto(0, 250)
pen.write("Score : 0", align="left", font=("Times", 30, "bold"))


def goUp():
    if head.direction != "down":
        head.direction = "up"
def goDown():
    if head.direction != "up":
        head.direction = "down"
def goLeft():
    if head.direction != "right":
        head.direction = "left"
def goRight():
    if head.direction != "left":
        head.direction = "right"
    
def move():
    if head.direction == "up":
        y = head.ycor()
        head.sety(y+20)
    if head.direction == "down":
        y = head.ycor()
        head.sety(y-20)
    if head.direction == "left":
        x = head.xcor()
        head.setx(x-20)
    if head.direction == "right":
        x = head.xcor()
        head.setx(x+20)
 
 
screen.listen()
screen.onkey(goUp, "Up")
screen.onkey(goDown, "Down")
screen.onkey(goLeft, "Left")
screen.onkey(goRight, "Right")

segments = []
 
# Main Gameplay
count = 0
while True:
    screen.update()
       
    if score == 2:
        img1 = Image.open('won.jpg')
        img1.show()
        img3 = Image.open('final.jpg')
        img3.show()
        
    if score == -2:
        img2 = Image.open('lost.jpg')
        img2.show()
        img3 = Image.open('final.jpg')
        img3.show()
               
    if head.xcor() > 290 or head.xcor() < -290 or head.ycor() > 290 or head.ycor() < -290 or score == 2 or score == -2:
        time.sleep(1)
        head.goto(0, 0)
        head.direction = "Stop"
        #colors = random.choice(['red', 'blue', 'green'])
        #shapes = random.choice(['square', 'circle'])
        for segment in segments:
            segment.goto(1000, 1000)
        segments.clear()
        score = 0
        delay = 0.1
        pen.clear()
        pen.write("Score : 0", align="left", font=("Times", 30, "bold"))
        
    if head.distance(food) < 20:
        x = random.randint(-270, 270)
        y = random.randint(-270, 270)
        food.goto(x, y)
        if count > 1:
            w = random.randint(-270, 270)
            z = random.randint(-270, 270)
            trash.goto(w, z)
            
        count += 1
        
        # Adding segment
        new_segment = turtle.Turtle()
        new_segment.speed(0)
        new_segment.shape("square")
        new_segment.color("orange")
        new_segment.penup()
        segments.append(new_segment)
        delay -= 0.001
        score += 1
        
        pen.clear()
        pen.write("Score : {} ".format(score), align="left", font=("Times", 30, "bold"))
        
    if head.distance(trash) < 20:
        x = random.randint(-270, 270)
        y = random.randint(-270, 270)
        trash.goto(x, y)
        
        if count > 1:
            w = random.randint(-270, 270)
            z = random.randint(-270, 270)
            food.goto(w, z)
            
        count += 1
        
        # Adding segment
        new_segment = turtle.Turtle()
        new_segment.speed(0)
        new_segment.shape("circle")
        new_segment.color("orange")
        new_segment.penup()
        segments.append(new_segment)
        delay -= 0.001
        score -= 1
        pen.clear()
        pen.write("Score : {} ".format(score), align="left", font=("Times", 30, "bold"))
    
    
    # Checking for head collisions with body segments
    for index in range(len(segments)-1, 0, -1):
        x = segments[index-1].xcor()
        y = segments[index-1].ycor()
        segments[index].goto(x, y)
    if len(segments) > 0:
        x = head.xcor()
        y = head.ycor()
        segments[0].goto(x, y)
    move()
             
    for segment in segments:
        if segment.distance(head) < 20:
            time.sleep(1)
            head.goto(0, 0)
            head.direction = "stop"
            colors = random.choice(['red', 'blue', 'green'])
            shapes = random.choice(['square', 'circle'])
            for segment in segments:
                segment.goto(1000, 1000)
            segment.clear()
 
            score = 0
            delay = 0.1
            pen.clear()
            pen.write("Score : {} ".format(score), align="left", font=("Times", 30, "bold"))
   
    time.sleep(delay)

 
screen.mainloop()
```
